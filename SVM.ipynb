```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
df
```

```python
df['target'] = iris.target
```

```python
df
```

```python
iris.target_names
```

```python
df[df.target==1]
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1= df[df.target==1]
df2=df[df.target==2]
```

```python
plt.title('SVM_Classification')
plt.xlabel('sepal length')
plt.ylabel("sepal width")
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], color = 'g', marker="+")
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], color = 'b', marker="+")
```

```python
plt.title('SVM_Classification')
plt.xlabel('petal length')
plt.ylabel("petal width")
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], color = 'g', marker="+")
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], color = 'b', marker="+")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','flower_name'], axis = 'columns')
```

```python
y= df.target
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size = 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
